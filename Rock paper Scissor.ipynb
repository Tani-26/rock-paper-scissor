```python
import tkinter as tk
from tkinter import messagebox
import random

# Initialize scores
user_score = 0
computer_score = 0

# Function to handle user's choice
def play_game(user_choice):
    global user_score, computer_score
    
    # Map numbers to choices
    choices = {0: "Rock", 1: "Paper", 2: "Scissors"}
    
    # Generate computer's choice
    computer_choice = random.randint(0, 2)
    
    # Determine the winner
    if user_choice == computer_choice:
        result = "It's a tie!"
    elif (user_choice == 0 and computer_choice == 2) or \
         (user_choice == 1 and computer_choice == 0) or \
         (user_choice == 2 and computer_choice == 1):
        result = "You win!"
        user_score += 1
    else:
        result = "Computer wins!"
        computer_score += 1
    
    # Display the result and scores
    messagebox.showinfo("Result", f"Your choice: {choices[user_choice]}\nComputer's choice: {choices[computer_choice]}\n\n{result}\n\nUser Score: {user_score}\nComputer Score: {computer_score}")

# Create the main window
root = tk.Tk()
root.title("Rock-Paper-Scissors Game")

# Function to handle button clicks
def button_click(choice):
    play_game(choice)
button_style = {
    "font": ("Arial", 18, "bold"),  # Larger and bold font
    "bg": "#FF5733",  # Custom color for buttons
    "fg": "white",    # White text
    "relief": "raised",
    "bd": 0,          # No border
    "width": 20,      # Increased width
    "padx": 30,       # Horizontal padding
    "pady": 15,       # Vertical padding
    "highlightthickness": 0  # No highlight
}

# Create buttons for rock, paper, and scissors
rock_button = tk.Button(root, text="Rock", command=lambda: button_click(0), **button_style)
rock_button.pack(pady=10)

paper_button = tk.Button(root, text="Paper", command=lambda: button_click(1), **button_style)
paper_button.pack(pady=10)
scissors_button = tk.Button(root, text="Scissors", command=lambda: button_click(2), **button_style)
scissors_button.pack(pady=10)

root.mainloop()
```
